```python
class People(object):
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

    def drive(self):
        print('{},{}岁,{},开车去广州'.format(self.name, self.age, self.gender))

    def sport(self):
        print('{},{}岁,{},最喜欢运动'.format(self.name, self.age, self.gender))

    def work(self):
        print('{},{}岁,{},去公司上班'.format(self.name, self.age, self.gender))


p1 = People('小明', '18', '男')
p2 = People('小红','28','女')
p3 = People('小东','22','男')
p1.drive()
p1.sport()
p1.work()

p2.drive()
p2.sport()
p2.work()

p3.drive()
p3.sport()
p3.work()
```

```python
class Role(object):
    def __init__(self, name, gender, age, vigor):
        self.name = name
        self.gender = gender
        self.age = age
        self.vigor = vigor

    def action(self):
        self.vigor -= 200
        print('草丛战斗，消耗200战斗力,还剩下{}战斗力'.format(self.vigor))

    def online(self):
        self.vigor -= 500
        print('多人游戏，消耗500战斗力，还剩下{}战斗力'.format(self.vigor))

    def retreat(self):
        self.vigor -= 100
        print('自我修炼，消耗战斗力100,还剩下{}战斗力'.format(self.vigor))


r1 = Role('苍井空', '女', 18, 2500)

r1.action()
r1.online()
r1.retreat()
```

```python
class Student(object):
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def get_name(self):
        return self.name

    def get_age(self):
        return self.age

    def get_course(self):
        grade = max(self.grade) #3门科目中最高的分数
        return grade


s1 = Student('小明', 18, [90, 80, 70])
print(s1.get_name())
print(s1.get_age())
print(s1.get_course())
```

```python

# 定义一个字典类：dictclass
class Dictclass(object):
    def del_dict(self, dict, key): #删除某个key :del_dict(key) 
        dict.pop(key)
        return dict

    def get_dict(self, dict, key):# 如果在返回键对应的值，不存在则返回"not found":get_dict(key)
        if key in dict:
            return dict[key]
        else:
            return 'not found'

    def get_key(self, dict):# 返回键组成的列表：返回类型;(list) 
        return list(dict)

    def update_dict(self, dic, dic1): #合并字典，并且返回合并后字典的values组成的列表。返回类型:(list)
     
        dic = dict(dic, **dic1)
        return list(dic.values())


d1 = Dictclass()
dic = {1: 'aa', 2: 'bb', 3: 'cc', 4: 'dd'}

print(d1.del_dict(dic, 1))
print(d1.get_dict(dic, 2))
print(d1.get_key(dic))
print(d1.update_dict(dic, {'a': 'aa', 'b': '22'}))
```

```python
 # NumPy(Numerical Python) 是 Python 语言的一个扩展程序库,支持大量的维度数组与矩阵运算
 # linspace(start,stop,num=50endpoint=True,) (起始点,终止点,默认50，生成start和stop之间50个等差间隔的元素)

# setup
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-5,5,100)
f = np.sin(x)
# print(x)
# print(f)
```

# MATLAB style interface "MATLAB 风格的界面"

```python
# fig = plt.figure(figsize=(a, b), dpi=dpi)
# figsize 设置图形的大小，a 为图形的宽， b 为图形的高，单位为英寸
# dpi 为设置图形每英寸的点数

plt.figure(dpi = 100)
plt.plot(x,f)
plt.title("sin(x)") #标题
plt.xlabel("x") #X轴的名称
plt.ylabel("y") #Y轴的名称
# plt.show() # needed for .py-script
```

# OOP approach 方法

```python
# OOP approach 方法
# fig is the canvas where you can put draw one or several axes fig 是画布，您可以在其中绘制一个或多个轴
# an axes can contain axis, xlabel, ylabel, title, plot, scatter, histogram, bar, ..轴可以包含轴、xlabel、ylabel、标题、绘图、散点图、直方图、条形图、..
```

```python
# instantiates the objects fig and ax from Figure and AxesSubplot 
# 从 Figure 和 AxesSubplot 实例化对象 fig 和 ax

fig, ax = plt.figure(dpi=100),plt.axes()
ax.plot(x, f) # the ax object has a plot
# calling the method set from ax object to set the axis, labels and title 调用从 ax 对象设置的方法来设置轴、标签和标题

ax.set(xlim = (-6,6), ylim = (-1.5,1.5), xlabel="x", ylabel="y", title="$\sin{x}$");
# plt.show() # needed for .py-script
```

```python
import matplotlib.pyplot as plt 
fig, ax = plt.subplots(1, 2, dpi=150, figsize=(10, 4))
rng = np.random.RandomState(42)

# randn - draws random number from normal distribution
data_normal_dist = rng.randn(1000)

ax[0].hist(data_normal_dist, bins=30, alpha=.5) #ax[0]是第一个图，ax[1]是第二个图
ax[0].set(title="Histogram of normal distributed data", xlabel="x", ylabel="y")

bubble=ax[1].scatter(x=rng.randn(100),y=rng.randn(100),c=rng.randn(100),alpha=.7,cmap="cividis",s=1000*rng.rand(100)) 
ax[1].set(title="Bubble Chart", xlabel="x", ylabel="y")
fig.colorbar(bubble)

plt.savefig("beautiful_plots.png",facecolor="white")
```

```python
fig, ax = plt.subplots(1, 2, dpi=150, figsize=(10, 4))

rng = np.random.RandomState(42)
# randn - draws random number from normal distribution 从正态分布中抽取随机数
data_normal_dist = rng.randn(1000)

# 方柱图
ax[0].hist(data_normal_dist, bins=30, alpha=.5) #ax[0]是第一个图
ax[0].set(title="Histogram of normal distributed data", xlabel="x", ylabel="y")

#散点图
size = 1000*rng.rand(100)
sc = ax[1].scatter(x=rng.randn(100), y=rng.randn(100), c= rng.randn(100), alpha=.3, cmap="viridis", s=size)  #cmap=热力图颜色
fig.colorbar(sc)
ax[1].set(title="Bubble chart", xlabel="x", ylabel="y") 

plt.savefig("../assets/beautiful_plots.png", transparent=True, facecolor="white")  #transparent 透明的
# plt.show() # needed for .py-script
```
