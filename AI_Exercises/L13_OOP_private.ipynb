```python
# setup
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-5,5,100)
f = np.sin(x)
print(x)
print(f)
```

```python
# MATLAB style interface

# fig = plt.figure(figsize=(a, b), dpi=dpi)
# figsize 设置图形的大小，a 为图形的宽， b 为图形的高，单位为英寸
# dpi 为设置图形每英寸的点数

plt.figure(dpi = 100)
plt.plot(x,f)
plt.title("sin(x)") #标题
plt.xlabel("x") #X轴的名称
plt.ylabel("y") #Y轴的名称
# plt.show() # needed for .py-script
```

```python
# OOP approach 方法
# fig is the canvas where you can put draw one or several axes fig 是画布，您可以在其中绘制一个或多个轴
# an axes can contain axis, xlabel, ylabel, title, plot, scatter, histogram, bar, ..轴可以包含轴、xlabel、ylabel、标题、绘图、散点图、直方图、条形图、..
```

```python
# instantiates the objects fig and ax from Figure and AxesSubplot 从 Figure 和 AxesSubplot 实例化对象 fig 和 ax
fig, ax = plt.figure(dpi=100),plt.axes()
ax.plot(x, f) # the ax object has a plot
# calling the method set from ax object to set the axis, labels and title 调用从 ax 对象设置的方法来设置轴、标签和标题
ax.set(xlim = (-6,6), ylim = (-1.5,1.5), xlabel="x", ylabel="y", title="$\sin{x}$");
# plt.show() # needed for .py-script
```

```python
fig, ax = plt.subplots(1, 2, dpi=150, figsize=(10, 4))
rng = np.random.RandomState(42)
data_normal_dist = rng.randn(1000)

ax[0].hist(data_normal_dist, bins=30, alpha=.5) #ax[0]是第一个图，ax[1]是第二个图
ax[0].set(title="Histogram of normal distributed data", xlabel="x", ylabel="y")

bubble=ax[1].scatter(x=rng.randn(100),y=rng.randn(100),c=rng.randn(100),alpha=.7,cmap="cividis",s=1000*rng.rand(100)) 
ax[1].set(title="Bubble Chart", xlabel="x", ylabel="y")
fig.colorbar(bubble)

plt.savefig("beautiful_plots.png",facecolor="white")
```

```python
fig, ax = plt.subplots(1, 2, dpi=150, figsize=(10, 4))

rng = np.random.RandomState(42)
# randn - draws random number from normal distribution 从正态分布中抽取随机数
data_normal_dist = rng.randn(1000)

# 方柱图
ax[0].hist(data_normal_dist, bins=30, alpha=.5) #ax[0]是第一个图
ax[0].set(title="Histogram of normal distributed data", xlabel="x", ylabel="y")

#散点图
size = 1000*rng.rand(100)
sc = ax[1].scatter(x=rng.randn(100), y=rng.randn(100), c= rng.randn(100), alpha=.3, cmap="viridis", s=size)  #cmap=热力图颜色
fig.colorbar(sc)
ax[1].set(title="Bubble chart", xlabel="x", ylabel="y") 

plt.savefig("../assets/beautiful_plots.png", transparent=True, facecolor="white")  #transparent 透明的
# plt.show() # needed for .py-script
```
